```python
import cv2
import time

# Load video
video_path = r"C:\Users\tsion\Downloads\cv-lab2-opencv-ml\videos\city_traffic.mp4"
cap = cv2.VideoCapture(video_path)

# Load OpenCV's built-in Haar cascade face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Get video properties
fps_video = cap.get(cv2.CAP_PROP_FPS)
frame_limit = int(fps_video * 10)

# Save output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('haar_face_output_10s.mp4', fourcc, fps_video, (int(cap.get(3)), int(cap.get(4))))

start_time = time.time()
frame_count = 0

while cap.isOpened() and frame_count < frame_limit:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    out.write(frame)
    cv2.imshow("Haar Cascade Face Detection", frame)
    frame_count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

fps = frame_count / (time.time() - start_time)
print(f"Haar Cascade FPS: {fps:.2f}")
```

```python
import cv2
import numpy as np
import time

# Load video
video_path = r"C:\Users\tsion\Downloads\cv-lab2-opencv-ml\videos\city_traffic.mp4"
cap = cv2.VideoCapture(video_path)

# Load YOLOv8 ONNX model
net = cv2.dnn.readNetFromONNX('yolov8n.onnx')

# Get video properties
fps_video = cap.get(cv2.CAP_PROP_FPS)
frame_limit = int(fps_video * 10)

# Save output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('yolo_onnx_output_10s.mp4', fourcc, fps_video, (int(cap.get(3)), int(cap.get(4))))

start_time = time.time()
frame_count = 0
CONF_THRESHOLD = 0.5
NMS_THRESHOLD = 0.4

while cap.isOpened() and frame_count < frame_limit:
    ret, frame = cap.read()
    if not ret:
        break

    h, w = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (640, 640), swapRB=True, crop=False)
    net.setInput(blob)
    outputs = net.forward()

    # Post-processing: extract boxes, scores, class IDs
    boxes = []
    confidences = []
    class_ids = []

    for detection in outputs[0, 0]:
        confidence = detection[4]
        if confidence > CONF_THRESHOLD:
            scores = detection[5:]
            class_id = np.argmax(scores)
            score = scores[class_id]
            if score > CONF_THRESHOLD:
                cx, cy, w_box, h_box = detection[:4]
                x = int((cx - w_box / 2) * w)
                y = int((cy - h_box / 2) * h)
                width = int(w_box * w)
                height = int(h_box * h)
                boxes.append([x, y, width, height])
                confidences.append(float(score))
                class_ids.append(class_id)

    indices = cv2.dnn.NMSBoxes(boxes, confidences, CONF_THRESHOLD, NMS_THRESHOLD)

    for i in indices:
        i = i[0]
        x, y, width, height = boxes[i]
        cv2.rectangle(frame, (x, y), (x + width, y + height), (0, 255, 0), 2)
        cv2.putText(frame, f"Class {class_ids[i]}", (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    out.write(frame)
    cv2.imshow("YOLOv8 ONNX Detection", frame)
    frame_count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()

fps = frame_count / (time.time() - start_time)
print(f"YOLOv8 ONNX FPS: {fps:.2f}")
```
